```python
def get_genres(isbn):
    try:
        response = requests.get(f'https://www.goodreads.com/book/isbn/{isbn}')
        soup = BeautifulSoup(response.content, 'html.parser')
        genre_tags = soup.select("#__next > div:nth-of-type(2) > main > div:first-of-type > div:nth-of-type(2) > div:first-of-type > div:nth-of-type(2) > div:nth-of-type(5) > ul > span:first-of-type > span > a > span")
        genres = [tag.text.strip() for tag in genre_tags]
    except requests.exceptions.RequestException:
        genres = []
    return genres
```

```python
df['genres'] = df['isbn'].apply(get_genres)
```
